```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
matplotlib.style.use('ggplot')

import pandas as pd
```

```python
exel = pd.ExcelFile("/home/awinhold/Desktop/filter-bandpasses-9_2_16.xlsx")
```

```python
# Confirm the name of the excel sheet
exel.sheet_names
```

```python
df = exel.parse("filter testing 9/2/16", index='Oversized Cut, Wavelength (nm)')
```

```python
df.head()
```

```python
# Remove the empty columns
df.drop('Unnamed: 3', axis=1, inplace=True)
df.drop('Unnamed: 7', axis=1, inplace=True)
```

```python
df.head()
```

```python
# Rename the column headers
df.columns = ['oversized', 'oversized-perp', 'oversized-angle', 'regular', 'reg-perp','reg-angle','depths','depths-perp','depths-angle']
df.head()
```

```python
# Drop the "80th" (really 82nd) row because it makes the data display a little rough, its redundant if you look at the data.
df.drop(df.index[80], inplace=True)
df.head(n=82)
```

```python
# Oversized
fig, ax = plt.subplots(1, figsize=(18,9))
plt.xticks(np.arange(min(df['regular']), max(df['regular']), 20))
ax.plot(df['oversized'], df['oversized-perp'], marker='.')
ax.plot(df['oversized'], df['oversized-angle'], marker='.')
ax.set_title('Oversized Filter: Detector Perpendicular and Detector at Angle')
ax.legend(loc=1)
ax.set_xlabel('Wavelength (nm)')
ax.set_ylabel('Optical Power (nW)')
fig.savefig('oversized-filter-300-1100-period.png')
```

```python
# Oversized, 1nm
fig, ax = plt.subplots(1, figsize=(18,9))
plt.xticks(np.arange(min(df['regular']), max(df['regular']), 5))
ax.plot(df['oversized'][30:82], df['oversized-perp'][30:82], marker='o')
ax.plot(df['oversized'][30:82], df['oversized-angle'][30:82], marker='o')
ax.set_title('Oversized Filter: Detector Perpendicular and Detector at Angle, 1nm steps')
ax.legend(loc=1)
ax.set_xlabel('Wavelength (nm)')
ax.set_ylabel('Optical Power (nW)')
fig.savefig('oversized-filter-575-650.png')
```
